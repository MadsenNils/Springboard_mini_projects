```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```

___
### Data Inspection
___

```python
# load data
logins = pd.read_json('../data/logins.json')
```

```python
logins.head()
```

```python
# there are no missing values
logins.info()
```

```python
# datetime range of logins
login_range = (logins.login_time.min(), logins.login_time.max())
print(login_range)
```

___
This dataset is limited in the fact that it only includes login timestamps from the first 3.5 months of the year. This makes it impossible to fully explore the trends of logins across different months of the year. Certain important, high-traffic periods of time, such as the holidays in November and December, have no data to analyze. A dataset including a full 12-month period would be more useful and reveal more insights into login trends.

The 1st of January and 13th of April also look to be incomplete days, so they will be removed.
___

```python
# remove Jan 1
logins = logins[logins.login_time >= pd.Timestamp(1970, 1, 2, 0)]

# remove April 13
logins = logins[logins.login_time < pd.Timestamp(1970, 4, 13, 0)]

# new datetime range
login_range = (logins.login_time.min(), logins.login_time.max())
print(login_range)
```

___
### Feature Extraction
___

```python
# extract month, day, day of week, hour, and minute from login_time
logins['month'] = logins.login_time.dt.month
logins['day'] = logins.login_time.dt.day
logins['day_of_week'] = logins.login_time.dt.dayofweek
logins['time'] = logins.login_time.dt.time

# set index to time of day
logins.set_index(pd.DatetimeIndex(pd.to_datetime(logins.time.astype(str))), inplace=True)
logins['time'] = logins.index

# number of days in date range
n_days = login_range[1].round('D') - login_range[0].round('D')
n_days = n_days.days #timedelta to int

# count for each weekday in date range
n_weeks = n_days//7
days_of_week = {i:n_weeks for i in range (7)}
if login_range[1].weekday() > login_range[0].weekday():
    end_date = login_range[1].weekday()
else:
    end_date = login_range[1].weekday() +7
    
for i in range (login_range[0].weekday(), end_date + 1):
    days_of_week[i %7] +=1
```

___
### Data Exploration

___

```python
# logins per day by time of day (bins of 15 mins)
by_time = logins.resample('15min').login_time.count()/n_days

# convert index to normal Python datetime, for use in matplotlib formatter
by_time.index = by_time.index.to_pydatetime()
```

```python
# plot
fig, ax = plt.subplots(figsize=(8, 7))
ax.plot(by_time.index, by_time.values, color='black')

plt.title('Logins by Time')
plt.ylabel('Logins per Day')
plt.xlabel('Time (bins of 15 min)')
plt.xlim(by_time.index[0],by_time.index[-1])
plt.ylim(0, 20)

# x-axis ticks every 3 hours
ax.xaxis.set_major_locator(mdates.HourLocator(interval=3))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
fig.autofmt_xdate()
plt.xticks(rotation=30, horizontalalignment='center')
ax.xaxis.grid(True)

plt.show()
```

___
Logins peak at around noon and midnight, with large dropoffs in usage between these peaks.
___

```python
# login count by time and weekday
by_wd = logins.groupby('day_of_week').login_time.count()

# divide by count of each weekday
by_wd /= pd.Series(days_of_week)
```

```python
# plot
fig, ax = plt.subplots(figsize=(8, 7))
ax.plot(by_wd.index, by_wd.values, color='black')
plt.title('Logins by Day of Week')
plt.ylabel('Logins per Day')
plt.xlabel('Day of Week')
plt.ylim(0, 1350)
ax.set_xticklabels(['', 'Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])

plt.show()
```

___
Usage shows an increasing trend throughout the week, with Monday showing the fewest logins per day on average, and Saturday and Sunday showing the most logins.
___

```python
# login count by time and weekday
by_wd_time = logins.groupby(['day_of_week', pd.Grouper(key='time', freq='15min')]).login_time.count()
by_wd_time = by_wd_time.unstack(level=0)

# divide by count of each weekday
for col in by_wd_time.columns:
    by_wd_time[col] /= days_of_week[col]
    
by_wd_time.columns = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']

# convert index to normal Python datetime, for use in matplotlib formatter
by_wd_time.index = by_wd_time.index.to_pydatetime()
```

```python
# plot
fig, ax = plt.subplots(figsize=(8, 7))

for col in by_wd_time.columns:
    ax.plot(by_wd_time.index, by_wd_time[col])

plt.title('Logins by Time by Day of Week')
plt.ylabel('Logins per Day')
plt.xlabel('Time (bins of 15 min)')
plt.xlim(by_time.index[0],by_time.index[-1])
plt.ylim(0, 45)

# x-axis ticks every 3 hours
ax.xaxis.set_major_locator(mdates.HourLocator(interval=3))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
fig.autofmt_xdate()
plt.xticks(rotation=30, horizontalalignment='center')
ax.xaxis.grid(True)
ax.legend()

plt.show()
```

___
Separating the time of day trend by day of the week reveals different usage patterns between weekdays and weekends. Saturday and Sunday show a distinct peak around 4:30am, while there is no such peak on the weekdays. In fact, this peak is the highest-usage time period for the whole week.

On the other side of the coin, weekdays show a distinct peak around noon, while on weekends the usage shows a more gradual rise until around 3:00pm. 

The peak around 10:30pm is evident on all days except Sunday.
