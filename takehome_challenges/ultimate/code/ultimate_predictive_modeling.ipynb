```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
from scipy.stats import ttest_ind, skew

sns.set(style="whitegrid")
```

```python
# load dataset
filepath = '../data/ultimate_data_challenge.json'

with open(filepath) as f:
    data = json.load(f)

data = pd.DataFrame(data)

# convert date columns to datetime objects
data['last_trip_date'] = pd.to_datetime(data.last_trip_date)
data['signup_date'] = pd.to_datetime(data.signup_date)

# convert bool column to string
data['ultimate_black_user'] = data['ultimate_black_user'].astype(str)
```

___
### Data Cleaning
___

```python
data.info()
```

```python
data.phone.value_counts()
```

___
The missing values in the phone field can be filled with a placeholder category. This will allow ML algorithms to use these observations without changing the data.

The missing values in the rating_by_driver and rating_of_driver fields can be filled with the median of the respective field.
___

```python
# fill missing values in phone field with placeholder category
data['phone'] = data.phone.fillna('Unknown')

# fill missing values in remaining fields with median
data = data.fillna(data.median())
```

### Extracting the target variable
___

A user is defined as 'retained' if they took a trip in the last 30 days.

```python
# extract target variable
data['retained'] = 0
mask = data.last_trip_date >= data.last_trip_date.max() - pd.Timedelta('30 days')
data.loc[mask, 'retained'] = 1
```

#### Relationship of the numeric variables with the target

```python
def test_and_plot(data, field, target):
    '''
    Runs two-tailed t-test to determine whether a numeric field is significantly 
        different between retained and non-retained users
    Also creates box plot to visualize the relationship
    '''
    
    print('{}:'.format(field))
    
    # run t-test with null hypothesis of equal means
    positive = data[data[target] == 1]
    negative = data[data[target] == 0]
    tstat, pvalue = ttest_ind(positive[field], negative[field], equal_var=True)
    
    # box plot
    sns.boxplot(x=target, y=field, data=data)
    plt.show()
    
    print('positive mean = {}'.format(positive[field].mean()))
    print('negative mean = {}'.format(negative[field].mean()))
    print('tstat = {}'.format(tstat))
    print('pvalue = {}'.format(pvalue))
    print('\n')
    
    return None
```

```python
for field in numeric_fields:
    test_and_plot(data, field, target='retained')
```

___
All of the numeric variables except avg_surge are significantly different (alpha=0.05) between the retained and unretained users. For most of these variables, the actual difference between means is small. One notable difference between the two groups is the large difference in mean trips in the first 30 days, which is three times higher in the retained group.

The boxplots reflect the skewed nature of the distributions, showing the many outliers present in the data.
___

#### Relationship of the categorical variables with the target

```python
for field in category_fields:
    # plot proportion of users retained for each class in the field
    sns.barplot(x=field, y='retained', data=data, capsize=0.05)
    plt.title('Retention proportion by {}'.format(field))
    plt.show()
    
    # perform proportions z-test for each pair of classes in the field
    from itertools import combinations
    from statsmodels.stats.proportion import proportions_ztest
    
    combs = [(x[0], x[1]) for x in combinations(data[field].unique(), 2)]
    
    for c in combs:
        npos = [data[data[field] == x].retained.sum() for x in c]
        nobs = [data[data[field] == x][field].count() for x in c]
        zstat, pvalue = proportions_ztest(npos, nobs)
        
        print('{} vs {} p-value: {}'.format(c[0], c[1], pvalue))
    
    print('\n')
```

___
Within the city field, users in King's Landing had the highest retention, while users in Astapor had the lowest retention.

Within the phone field, users with an iPhone were retained better than users with an Android phone. Users who had missing data in this field had a retention between iPhone and Android users.

Ultimate black users were retained better than non-ultimate-black users.

All classes in all categories were significantly different from each other (alpha=0.05) based on proportions z-test.

### Predictive Modeling
___

#### Model selection

Random forest and logistic regression are both powerful algorithms for binary classification with a mixture of categorical and numeric fields. The categorical fields in this dataset all have a low number of classes (2-3), so implementations of these two models that use one-hot encoding for categorical features (scikit-learn) will be capable of performing well on this task. If the categorical fields had more classes, classification algorithms that work directly with the categorical features (i.e. without one-hot encoding), such as Light-GBM or H2O.ai's implementation of random forest, would be capable of much better performance.

Random forest and logistic regression also allow for intuitive interpretation of the relative importance of each feature in the prediction, allowing better understanding of how to utilize the results of the models.

#### Dividing the data into training and test sets

```python
from sklearn.model_selection import train_test_split

X = data.copy().drop(['retained', 'last_trip_date', 'signup_date'], axis=1)

y = data.retained

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=101)
```

```python
X_train.info()
```

#### Un-skewing the numeric fields

Logistic regression generally performs better on features that are not skewed.

```python
skewed_fields = numeric_fields.copy()
skewed_fields.remove('weekday_pct') # don't un-skew a bimodal distribution

# add a small constant value to make input positive
k = 0.1

for field in ['avg_dist', 'surge_pct', 'trips_in_first_30_days']:
    X_train[field] += k
    X_test[field] += k
```

```python
for field in skewed_fields:
    from scipy.stats import boxcox
    
    # apply Box-Cox power transform to unskew distribution
    transformed, _ = boxcox(X_train[field])
    
    # plot transformed distribution
    print('{}:'.format(field))
    sns.distplot(transformed)
    plt.show()
    print('skew: {}\n'.format(skew(transformed)))
```

___
Only avg_dist can be unskewed to a satisfactory degree. The other fields separate into discrete values when unskewed and will have to be left as-is.
___

```python
# apply Box-Cox power transform to unskew avg_dist
X_train['avg_dist'], lmdba = boxcox(X_train['avg_dist'])
X_test['avg_dist'] = boxcox(X_test['avg_dist'], lmbda=lmdba)

# restore original values in other fields
for field in ['surge_pct', 'trips_in_first_30_days']:
    X_train[field] -= k
    X_test[field] -= k
```

```python
# convert categorical fields to dummy variables
X_train['ultimate_black_user'] = X_train.ultimate_black_user.astype(bool).astype(np.uint8)
X_train = pd.get_dummies(X_train)
X_train.drop('phone_Unknown', axis=1, inplace=True)

X_test['ultimate_black_user'] = X_test.ultimate_black_user.astype(bool).astype(np.uint8)
X_test = pd.get_dummies(X_test)
X_test.drop('phone_Unknown', axis=1, inplace=True)
```

#### Feature Selection

```python
# standardize features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_scaled = scaler.fit_transform(X_train)
```

```python
# get baseline performance of random forest and logistic regression
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

forest = RandomForestClassifier(n_estimators=100,
                                min_samples_leaf=5,
                                max_features=None,
                                random_state=42,
                                n_jobs=7)

forest_score = cross_val_score(forest, train_scaled, y_train)

logit = LogisticRegression(penalty='l2',
                           C=1,
                           random_state=42)

logit_score = cross_val_score(logit, train_scaled, y_train)
```

```python
print('forest acc: {}'.format(forest_score.mean()))
print('logit acc: {}'.format(logit_score.mean()))
```

___
Random forest performs better at baseline than logistic regression. This could be due to a combination of 1) random forest does not suffer from high dimensionality, and 2) random forest does not suffer from skewed feature distributions, both of which are important aspects of this dataset.

Reducing the dimensionality of the dataset might improve the performance of the logistic model.
___

```python
# plot feature importances for random forest
forest.fit(train_scaled, y_train)
rf_features = pd.Series(forest.feature_importances_, index=X_train.columns)
rf_features.sort_values(inplace=True)
rf_features.plot(kind='barh')
```

```python
# plot feature importances for logistic regression
logit.fit(train_scaled, y_train)
logit_features = pd.Series(np.absolute(logit.coef_[0]), index=X_train.columns)
logit_features.sort_values(inplace=True)
logit_features.plot(kind='barh')
```

___
The two models are quite different in their judgment of feature importances.
___

```python
# remove lowest-importance features for logistic regression
train_trimmed = X_train.drop(['ultimate_black_user', 'weekday_pct', 'avg_dist'], axis=1)
train_trim_scaled = scaler.fit_transform(train_trimmed)
```

```python
# reevaluate performance of models
forest_score = cross_val_score(forest, train_trim_scaled, y_train)
logit_score = cross_val_score(logit, train_trim_scaled, y_train)

print('forest acc: {}'.format(forest_score.mean()))
print('logit acc: {}'.format(logit_score.mean()))
```

___
Reducing the dimensionality of the dataset did not improve the performance of the logistic model, and hurt the performance of the random forest model. This likely indicates a fundamental limitation of logistic regression on this dataset, perhaps related to the degree of skew in the feature distributions. Random forest will be the only model considered from this point on.

Since random forest is not affected by standardization of features, standard scaler will not be used.
___

#### Hyperparameter tuning

```python
# tune random forest hyperparameters
from sklearn.model_selection import GridSearchCV

param_grid = {'max_features': ['auto', None],
              'min_samples_leaf': [1, 3, 5, 7, 9]}

forest = RandomForestClassifier(n_estimators=100,
                                random_state=42)

forest_grid = GridSearchCV(forest, param_grid, cv=5, n_jobs=7)

forest_grid.fit(X_train, y_train)

print(forest_grid.best_params_)
print(forest_grid.best_score_)
```

```python
# tune number of trees
param_grid = {'n_estimators': range(100,1100, 100)}

forest = RandomForestClassifier(max_features='auto',
                                min_samples_leaf=7,
                                random_state=42)

forest_grid = GridSearchCV(forest, param_grid, cv=5, n_jobs=7)

forest_grid.fit(X_train, y_train)

print(forest_grid.best_params_)
print(forest_grid.best_score_)
```

___
The best hyperparameters found for this model were max_features set to the square root of the number of features in the dataset ('auto') and a minimum of 7 samples per leaf of the tree. Performance of the model did not improve by increasing the number of trees above 100.
___

#### Model Evaluation

```python
# fit best model
model = forest_grid.best_estimator_
model.fit(X_train, y_train)
```

```python
# evaluate model on the test set
from sklearn.metrics import roc_auc_score

probs = model.predict_proba(X_test)[:,1]

print('accuracy: {}'.format(model.score(X_test, y_test)))
print('AUC score: {}'.format(roc_auc_score(y_test, probs)))
```

```python
# plot model success metrics by threshold
from sklearn.metrics import confusion_matrix

def threshold_func_gen(threshold):
    ''' creates a thresholding function to convert probabilities to class designation '''
    def threshold_func(value):
        if value >= threshold:
            return 1
        else:
            return 0
    
    return threshold_func

predicted_probs = pd.Series(model.predict_proba(X_test)[:,1])
threshold_options = list(np.linspace(0, 1.0, num=101))
sens_by_thresh = np.empty(len(threshold_options))
spec_by_thresh = np.empty(len(threshold_options))
acc_by_thresh = np.empty(len(threshold_options))

for i, threshold in enumerate(threshold_options):
    threshold_func = threshold_func_gen(threshold)
    y_pred = predicted_probs.map(threshold_func)
    matrix = confusion_matrix(y_test, y_pred)
    sens_by_thresh[i] = matrix[1, 1]/np.sum(matrix[1,:])
    spec_by_thresh[i] = matrix[0, 0]/np.sum(matrix[0,:])
    acc_by_thresh[i] = (matrix[0,0] + matrix[1,1])/np.sum(matrix)
    
threshold_df = pd.DataFrame(threshold_options)
threshold_df['sensitivity'] = sens_by_thresh
threshold_df['specificity'] = spec_by_thresh
threshold_df['accuracy'] = acc_by_thresh
threshold_df.set_index(0, inplace=True)
threshold_df.index.name = 'threshold'
threshold_df.plot()
plt.title('Sensitivity, Specificity, and Accuracy by Threshold')
plt.xlabel('Threshold')
plt.show()
print(threshold_df)
```

___
The model shows decent performance at an accuracy of 77.4% and AUC score of 0.84 on the test set. Plotting performance metrics by threshold shows that altering the threshold of the model would not improve the overall accuracy. However, it might still be useful for tuning the model to preferentially detect either the negative or positive class, based on business needs.

___

One way this model could be leveraged to yield benefit for Ultimate is to use the prediction of the model to identify users that are most likely to stop using the service. Approaches such as customer engagement emails or promotions could then be targeted to try to reduce attrition in these users.

Beyond the prediction itself, the importance that the model attributes to each feature in the dataset can be used to glean possible approaches to improving user retention overall.


```python
# plot feature importances for random forest
model_features = pd.Series(model.feature_importances_, index=X_train.columns)
model_features.sort_values(inplace=True)
model_features.plot(kind='barh')
plt.show()
```

The most important feature, according to random forest, is the average rating of the user by drivers. Strangely, users that were not retained had higher average rating by drivers than those that were retained. Similarly, inactive users rated their drivers higher on average than did active users. This suggests that user attrition is not, in general, caused by poor interactions with drivers.

With regard to percent of trips taken during the weekday, retained users had a higher percent of weekday trips. This might indicate some interaction such as some customers using the service to get to work, which would naturally lead to high retention out of necessity. This warrants a closer exploration.
